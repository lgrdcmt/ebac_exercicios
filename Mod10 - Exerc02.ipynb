````python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
````

````python
df_raw = pd.read_csv('previsao_de_renda.csv')

df_raw.head()
````

````python
df = df_raw.copy()

df.drop(['Unnamed: 0', 'mau', 'index'], axis = 1, inplace = True)

df.head()
````

````python
#posse_de_imovel

ax = sns.pointplot(x = 'posse_de_imovel', y = 'renda', 
                   data = df, dodge = True, ci = 90)
````

````python
#posse_de_veiculo

ax = sns.pointplot(x = 'posse_de_veiculo', y = 'renda', 
                   data = df, dodge = True, ci = 90)
````

- A variável ``posse_de_veiculo``, aparenta ser mais signifficante que ``posse_de_imovel``. A média com relação a quem tem ``posse_de_veiculo`` é maior sobre a ``posse_de_imovel``.
