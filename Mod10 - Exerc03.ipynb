````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
````

````python
df_raw = pd.read_csv('Mod10 - previsao_de_renda.csv')
````

````python
df = df_raw.copy()

df.drop(['Unnamed: 0', 'index'], axis = 1, inplace = True)

df.head()
````

````python
df['data_ref'] = pd.to_datetime(df['data_ref'])

df.head()
````

````python
_, ax = plt.subplots(figsize=(20, 10))

ax = sns.countplot(x = 'data_ref',hue = 'mau', data = df)

tick_labs = df['data_ref'].map(lambda ts: ts.strftime("%m-%Y")).unique()

ticks = ax.set_xticks(list(range(df['data_ref'].nunique())))
labels = ax.set_xticklabels(tick_labs, rotation = 0)

plt.legend(bbox_to_anchor = (1.05, 1), loc = 2, borderaxespad = 0)
````

- Acredito, que não teve nenhuma alteração, pois em todos os meses se manteve parecido
