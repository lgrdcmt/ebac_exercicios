```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#%matplotlib notebook
```

```python
arquivo = 'Mod06 - Aula 01 CSV.csv'

sinasc_ro = pd.read_csv(arquivo)

sinasc_ro
```

####  2) Monte um gráficos de barras com as variáveis APGAR1, APGAR5, SEXO, GESTACAO, ESTCIVMAE.

```python
sinasc_ro[['APGAR1', 'APGAR5', 'SEXO', 'GESTACAO', 'ESTCIVMAE']].head()
```

```python
fig = plt.figure(figsize = (10, 10))

ax1 = fig.add_subplot(2, 2, 1) #APGAR1
ax1 = sinasc_ro['APGAR1'].value_counts().sort_index().plot.bar(xlabel = 'APGAR1')
ax2 = fig.add_subplot(2, 2, 2) #APGAR5
ax2 = sinasc_ro['APGAR5'].value_counts().sort_index().plot.bar(xlabel = 'APGAR5')
ax3 = fig.add_subplot(2, 2, 3) #ESTCIVMAE
ax3 = sinasc_ro['ESTCIVMAE'].value_counts().sort_index().plot.bar(xlabel = 'ESTCIVMAE')

plt.subplots_adjust(wspace = 0.25, hspace = 0.25)
```

```python
fig1 = plt.figure(figsize = (10, 10))

ax1 = fig1.add_subplot(2, 2, 1) #SEXO
ax1 = sinasc_ro['SEXO'].value_counts().sort_index().plot.bar(xlabel = 'SEXO')
ax2 = fig1.add_subplot(2, 2, 2) #GESTACAO
ax2 = sinasc_ro['GESTACAO'].value_counts().sort_index().plot.bar(xlabel = 'GESTACAO')

plt.subplots_adjust(wspace = 0.25, hspace = 0.25)
```

```python
#OUTRA FORMA
plt.close('all')

plt.rc('figure', figsize = (10, 10))

fig, axes = plt.subplots(2, 2)

sns.countplot(ax = axes[0, 0], x = 'APGAR1', data = sinasc_ro)
sns.countplot(ax = axes[0, 1], x = 'APGAR5', data = sinasc_ro)
sns.countplot(ax = axes[1, 0], x = 'SEXO', data = sinasc_ro)
sns.countplot(ax = axes[1, 1], x = 'ESTCIVMAE', data = sinasc_ro)

plt.subplots_adjust(wspace = 0.25, hspace = 0.25)
```

```python
plt.close('all')

sns.countplot(x = 'GESTACAO', data = sinasc_ro).tick_params(axis = 'x', rotation = 30)
```

```python
### OUTRA FORMA COM OS 4
plt.close('all')
lista = [['APGAR1', 'APGAR5'], ['SEXO', 'ESTCIVMAE']]

plt.close('all')
plt.rc('figure', figsize = (10, 10))
fig, axes = plt.subplots(2, 2)

for i in range(2):
    for j in range(2):
        sns.countplot(ax = axes[i, j], x = lista[i][j], data = sinasc_ro)
        
plt.subplots_adjust(wspace = 0.25, hspace = 0.25)
```

#### 3) Monte um gráfico de barras mostrando o APGAR5 médio por CONSULTAS

```python
plt.close('all')

sns.barplot(x = 'CONSULTAS', y = 'APGAR5', data = sinasc_ro)
```

```python
#OUTRA FORMA
plt.close('all')

consultas = sinasc_ro['CONSULTAS'].unique()
apgar = []

for i in consultas:
    apgar.append(sinasc_ro.loc[sinasc_ro['CONSULTAS'] == i, 'APGAR5'].mean())

teste = pd.DataFrame(apgar, index = consultas, columns = ['APGAR5'])


teste.sort_index().plot.bar(xlabel='CONSULTAS', ylabel='APGAR5')
```

####  4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS

```python
plt.close('all')

lista = [['SEXO', 'GESTACAO'], ['ESTCIVMAE', 'CONSULTAS']]

plt.close('all')
plt.rc('figure', figsize = (10, 10))
fig, axes = plt.subplots(2, 2)

for i in range(2):
    for j in range(2):
        sns.countplot(ax = axes[i, j], x = lista[i][j], data = sinasc_ro)
        if i == 0 and j == 1:
            axes[i, j].tick_params(axis = 'x', rotation = 90)

plt.subplots_adjust(wspace = 0.25, hspace = 1)
```
