### Analise as respostas encontradas, tire algum insight delas, conte pra gente algo encontrado nos dados. Algo que você julgue relevante e novo pra você.

Exemplo:
- Ah, descobri que a idade mediana das mulheres que deram a luz no ano de 2019 dos municipios x é maior que y.

```python
sinasc_teste = sinasc_raw.copy()

sinasc_teste = sinasc_teste.loc[sinasc_teste['munResNome'] != 'Município ignorado - RO', 
                                ['DTNASC','munResNome', 'GESTACAO','QTDFILVIVO','IDADEMAE', 'IDADEPAI', 'SEXO', 'APGAR5', 'PARTO', 'ESCMAE']]

sinasc_teste['DTNASC'] = pd.to_datetime(sinasc_teste['DTNASC'])

sinasc_teste['MESNASC'] = pd.to_datetime(sinasc_teste['DTNASC']).dt.month

sinasc_teste.head(5)
```

```python
sinasc_teste_filtrado = sinasc_teste.loc[(sinasc_teste['MESNASC'] == 3) 
                                         & ((sinasc_teste['munResNome'] == 'Porto Velho') 
                                            | (sinasc_teste['munResNome'] == 'Castanheiras'))]
```

```python
sinasc_teste_filtrado[['munResNome','IDADEPAI']].dropna().groupby('munResNome').agg({('Média', np.mean), ('Máximo', max), ('Mínimo', min)})
```

- Descobri que em Castanheiras, a amplitude da idade dos pais no mês de Março é menor. 

```python
sinasc_teste['GESTSAUDAVEL'] = 0

sinasc_teste.loc[(sinasc_teste['GESTACAO'] == '37 a 41 semanas') | (sinasc_teste['GESTACAO'] == '42 semanas e mais'), 'GESTSAUDAVEL'] = 1


sinasc_teste['CONTADOR'] = 1

sinasc_teste_filtrado = sinasc_teste[['munResNome','GESTSAUDAVEL','CONTADOR']].groupby('munResNome').agg(sum)

sinasc_teste_filtrado['PORCENTAGEM'] = (sinasc_teste_filtrado['GESTSAUDAVEL'] / sinasc_teste_filtrado['CONTADOR']) * 100
```

```python
sinasc_teste_filtrado.sort_values('PORCENTAGEM')
```

- Descobri que em 2019, Cabixi foi a cidade que teve a maior porcentagem de Gestação Saudável, a julgar que Gestação saudável é a Gestação acima de 37 semanas. E que Guajará-Mirim, teve a piór porcentagem durante o ano de 2019, ainda em relação a Gestação Saudável.

```python
sinasc_teste.loc[sinasc_teste['APGAR5'] >= 8, 'ASFIXIA'] = 'Asfixia Normal'
sinasc_teste.loc[(sinasc_teste['APGAR5'] >= 6) & (sinasc_teste['APGAR5'] <= 7), 'ASFIXIA'] = 'Asfixia Leve'
sinasc_teste.loc[(sinasc_teste['APGAR5'] >= 4) & (sinasc_teste['APGAR5'] <= 5), 'ASFIXIA'] = 'Asfixia Moderada'
sinasc_teste.loc[sinasc_teste['APGAR5'] <= 3, 'ASFIXIA'] = 'Asfixia Severa'
```

```python
sinasc_teste_filtrado = sinasc_teste[['munResNome','ASFIXIA']].groupby(['munResNome'])['ASFIXIA'].value_counts().unstack().fillna(0)

sinasc_teste_filtrado.head(5)
```

```python
sinasc_teste_filtrado['TOTAL'] = (sinasc_teste_filtrado['Asfixia Leve'] + sinasc_teste_filtrado['Asfixia Moderada'] 
                                  + sinasc_teste_filtrado['Asfixia Normal'] + sinasc_teste_filtrado['Asfixia Severa'])

sinasc_teste_filtrado['PORCENTAGEM ASFIXIA'] = ((sinasc_teste_filtrado['Asfixia Leve'] + sinasc_teste_filtrado['Asfixia Moderada']  + sinasc_teste_filtrado['Asfixia Severa']) / sinasc_teste_filtrado['TOTAL']) * 100

sinasc_teste_filtrado.sort_values(by = 'PORCENTAGEM ASFIXIA', ascending = False).head(5)
```

- Descobri que Nova União, em 2019, teve uma porcentagem maior de bebes que sofreram asfixia durante os 5 minutos.

```python
sinasc_teste_filtrado = sinasc_teste[['ESCMAE','PARTO']].groupby('ESCMAE')['PARTO'].value_counts().unstack()

sinasc_teste_filtrado.sort_values(by = 'ESCMAE')
```

- Descobri que acima de 8 anos de escolaridade da mãe, o parto aparenta ser mais comum o ``Cesáreo``
