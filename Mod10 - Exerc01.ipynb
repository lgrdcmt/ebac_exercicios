1) Avalie a estrutura de correlação das variáveis quantitativas através de um gráfico de "matriz de dispersão" e através da avaliação gráfica da matriz de correlações. Veja se você identifica algum padrão interessante ou que te faça sentido.

````python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
````

````python
df_raw = pd.read_csv('previsao_de_renda.csv')

df_raw.head()
````

````python
df = df_raw.copy()

df.drop(['Unnamed: 0', 'mau', 'index'], axis = 1, inplace = True)

#df.drop(['Unnamed: 0', 'index'], axis = 1, inplace = True)
````

````python
df.corr()
````

````python
#cmap = sns.diverging_palette(h_neg = 125, h_pos = 350, as_cmap = True, sep = 60, center = 'light')

# Adicionei o annot = True, pois ficaria melhor para verificar a relação das variáveis
# e preferi usar o cmap = 'Greens', por questão de visualização

sns.clustermap(df.corr(), figsize = (10, 10), annot = True, center = 0, cmap = 'Greens');
````

````python
#OUTRA FORMA DE VISUALIZAÇÃO
_, ax = plt.subplots(figsize=(10, 10))

ax = sns.heatmap(df.corr(), annot = True, cmap = 'Greens')
````

- Olhando pelo gráfico gerado, a relação mais forte é a de qtd_filhos com qt_pessoas_residencia
- Outra relação que é possível de verificar pelo gráfico é a relação de idade com tempo_emprego
- E tempo_emprego com renda, também tem uma relação próxima

4) Calcule o logaritmo da variável renda e repita a sequência de análise
